```python
import matplotlib.pyplot as plt
import numpy as np
```

Primero generamos una imagen que representa un escalon corta duraci贸n en una dimensi贸n y un escalon de mayor duraci贸n en la otra dimensi贸n

```python
# Blank canvas
f = np.zeros((30,30))
f[4:25, 12:17] = 1
plt.imshow(f, cmap='gray');
```

Podemos ver que la imagen contiene una barra blanca. Esto podemos pensarlos como una ventana rectangular de ancho 5 y alto 20 con una intensidad igual a 1.

```python
# Perform FFT
F = np.fft.fft2(f)
F = np.fft.fftshift(F)
plt.imshow(np.abs(F), cmap='jet');
```

Del espectro podemos ver como ocurre la superposici贸n de las dos Sinc a partir del espectro de los pulsos originales

```python
plt.imshow(np.log(np.abs(F)), cmap='jet');
```

```python
# Look at the phases
plt.imshow(np.angle(F), cmap='jet');
```
