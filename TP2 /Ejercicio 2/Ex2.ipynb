```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

## Disk Filter

Primero se pasa la imagen por un filtro Pasabajos, por lo tanto, en la imagen resultante solo deben quedar los cambios suaves.

```python
import skimage 

im = plt.imread('barbara.gif')  # Abrimos la imagen original

plt.figure(figsize= (10,10))    # La Mostramos en pantalla
plt.imshow(im);
plt.show()


circle = skimage.morphology.disk(5, dtype= np.float32)
h = [[0, 1/6.0, 0],[1/6.0,1/3.0,1/6.0],[0,1/6,0]]
h = h / np.sum(h)
h= np.array(h)

disk_filter = cv.filter2D(circle,ddepth= -1, kernel= h, borderType= cv.BORDER_CONSTANT)

disk_filter = disk_filter/np.sum(disk_filter)

im = cv.filter2D(im,ddepth= -1, kernel= disk_filter)

plt.figure(figsize= (10,10))
plt.imshow(im);
plt.show()
```

Se puede observar que el resultado es adecuado debido a que el patron de lineas que previamente se encontraba en el mantel practicamente desaparece, como todas las porciones de la imagen donde habia cambio abruptos de brillo.

## Unsharp Filter


Ahora se le aplica un filtro pasaaltos, por lo que deberian pasar los cambios abruptos de brillo y atenuarse los cambios lentos

```python
from skimage.filters import unsharp_mask

im = plt.imread('barbara.gif')

im = cv.cvtColor(im,cv.COLOR_BGRA2GRAY)
plt.figure(figsize= (13,13))
plt.imshow(im, cmap= 'gray');
plt.show()

#h = [[0, 1/6.0, 0],[1/6.0,1/3.0,1/6.0],[0,1/6,0]]
#h = h / np.sum(h)
im = unsharp_mask(im)

#im = cv.filter2D(im,ddepth= -1, kernel= h)

plt.figure(figsize= (13,13))
plt.imshow(im, cmap= 'gray');
plt.show()
```

Como se puede observar, el resultado es adecuado debido a que en este caso se realzan los patrones del mantel, y el pantalon, que son rayados. A su vez se puede notar que tambien se realza el ruido del sensor de la camara.
