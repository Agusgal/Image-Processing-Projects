## Muestreo y cuantizaci√≥n

```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
xray_original = plt.imread('/content/drive/Shareddrives/Imagenes/TP2 Grupo 2/xray.jpg')
print(xray_original.shape)
```

```python
# Convert to grayscale
xray_gray_original = cv.cvtColor(xray_original, cv.COLOR_BGR2GRAY)
print(xray_gray_original.shape)
plt.imshow(xray_gray_original, cmap='gray');
```

```python
# ¬øCual es el rango actual de la imagen?
print("max: ",np.max(xray_gray_original))
print("min: ",np.min(xray_gray_original))
```

```python
gray_levels = [256, 128,  64, 32, 16, 8, 4, 3]
new_images = []
for gray_level in gray_levels:
  bins = np.linspace(0, 255, gray_level)
  digitized_image = bins[np.digitize(xray_gray_original, bins, right=False)-1]
  new_images.append(digitized_image)
```

Las imagenes van teniendo cada vez menos niveles de gris que puedan usar.

```python
upper_row = np.hstack(new_images[0:4])
second_row = np.hstack(new_images[4:])
new_collage = np.vstack([upper_row, second_row])
plt.figure(figsize=(15,15))
plt.imshow(new_collage, cmap='gray', vmin=0, vmax=255)
```
