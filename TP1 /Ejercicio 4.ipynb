```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mp
import scipy as sp
```

```python
def create_img(central_pix, surrounding_pixs):
  im = np.ones((3,3)) * surrounding_pixs
  im[1,1] = central_pix
  return im
```

```python
norm = mp.colors.Normalize(vmin= 0.0, vmax= 255.0, clip= True)
```

```python
im1 = create_img(central_pix= 127, surrounding_pixs= 63)
im2 = create_img(central_pix= 127, surrounding_pixs= 223)

im = np.concatenate((im1,im2), axis= 1)
plt.imshow(im,cmap= 'gray', norm= norm)
plt.show()
```

## Diferentes pruebas

```python
im1 = create_img(central_pix= 127, surrounding_pixs= 0)
im2 = create_img(central_pix= 127, surrounding_pixs= 255)

im = np.concatenate((im1,im2), axis= 1)
plt.imshow(im,cmap= 'gray', norm= norm)
plt.show()
```
